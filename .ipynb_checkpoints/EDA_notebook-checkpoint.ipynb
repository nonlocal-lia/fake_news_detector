```python
import os
from os.path import exists
```

```python
os.listdir('./data/politifact/fake/') 
```

```python
pol_fake_df = pd.DataFrame()
for d in os.listdir('./data/politifact/fake/'):
    if exists('./data/politifact/fake/'+ d +'/news content.json'):
        with open('./data/politifact/fake/'+ d +'/news content.json') as json_data:
            data = json.load(json_data)
        if 'og' in data['meta_data'].keys():
            if 'description' in data['meta_data']['og'].keys():
                df = pd.DataFrame([[data['title'], data['text'], data['meta_data']['og']['description'], 'Fake']])
            else:
                df = pd.DataFrame([[data['title'], data['text'], np.NAN, 'Fake']])
        else:
            df = pd.DataFrame([[data['title'], data['text'], np.NAN, 'Fake']])
        pol_fake_df = pd.concat([pol_fake_df, df], ignore_index=True)
pol_fake_df.columns = ['title', 'text', 'description', 'label']
pol_fake_df.head()
```

```python
pol_fake_df.info()
```

```python
pol_fake_df.head(20)
```

```python
pol_reliable_df = pd.DataFrame()
for d in os.listdir('./data/politifact/real/'):
    if exists('./data/politifact/real/'+ d +'/news content.json'):
        with open('./data/politifact/real/'+ d +'/news content.json') as json_data:
            data = json.load(json_data)
        if 'og' in data['meta_data'].keys():
            if 'description' in data['meta_data']['og'].keys():
                df = pd.DataFrame([[data['title'], data['text'], data['meta_data']['og']['description'], 'Real']])
            else:
                df = pd.DataFrame([[data['title'], data['text'], np.NAN, 'Real']])
        else:
            df = pd.DataFrame([[data['title'], data['text'], np.NAN, 'Real']])
        pol_reliable_df = pd.concat([pol_reliable_df, df], ignore_index=True)
pol_reliable_df.columns = ['title', 'text', 'description', 'label']
pol_reliable_df.head()
```
